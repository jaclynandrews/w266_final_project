```python
import pandas as pd
from pathlib import Path

source_files = sorted(Path('./drive/MyDrive/w266_final_project/MODELS/retraining/').glob('*.csv'))

dataframes = []
for file in source_files:
    df = pd.read_csv(file) # additional arguments up to your need
    df['source'] = file.name
    dataframes.append(df)

all = pd.concat(dataframes)

all.to_csv( "./drive/MyDrive/w266_final_project/MODELS/retraining/combined_csv_names.csv", index=False, encoding='utf-8-sig')
```
